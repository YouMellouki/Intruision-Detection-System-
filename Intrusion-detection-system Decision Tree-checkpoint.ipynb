## Preparing  data

```python
from pyspark.mllib.regression import LabeledPoint
from numpy import array
```

```python
data = raw_data.map(lambda x: x.split(","))
data.take(2)
```

```python
test_data = test_raw_data.map(lambda x: x.split(","))
```

```python
#we select the categorical variables to see their types 
network_protocols = data.map(lambda x: x[1]).distinct().collect()#we get the types of protocols of this network traffic
network_services = data.map(lambda x: x[2]).distinct().collect()# select distinct network services
network_flags = data.map(lambda x: x[3]).distinct().collect()#select the connexion status 
classes=data.map(lambda x: x[41]).distinct().collect()#the classes of attakcs on the network traffic
##hint  :The feature flags in the KDD99 dataset refers to the connexion status in which:##
#   SF: Normal SYN/FIN completion
#   REJ: Connection rejected, Initial SYN elicited a RST in reply
#   S0: initial SYN seen but no reply
#   S1: connection established (SYN's exchanged), nothing further seen
#   S2: connection established, initiator has closed their side
#   S3: connection established, responder has closed their side
#   RSTO: Connection reset by the originator
#   RSTR: Connection reset by the responder
#   OTH: a state not contemplated here
```

```python
#types of the protocols
network_protocols
```

```python
#network services on the destination
network_services
```

```python
#connexion types
network_flags
```

```python
#classes of attacks
classes
```

```python
#we use the predifined function 'labeled points ' to  change categorical variables to numerical ones
def create_labeled_point(line_split):
    # 41=='number of features'
    clean_line_split = line_split[0:41]
    
    # convert protocol_variable to numeric categorical variable
    try: 
        clean_line_split[1] = network_protocols.index(clean_line_split[1])
    except:
        clean_line_split[1] = len(network_protocols)
        
    # convert service_variable to numeric categorical variable
    try:
        clean_line_split[2] = network_services.index(clean_line_split[2])
    except:
        clean_line_split[2] = len(network_services)
    
    # convert flag_variable to numeric categorical variable
    try:
        clean_line_split[3] = network_flags.index(clean_line_split[3])
    except:
        clean_line_split[3] = len(network_flags)
    
    # convert label to binary label 
    """we will group the classes of attacks into one class labeled 'attack' in order to get a binary classification 
    called attack or normal """
    #if label='attack' change to 1.0 else label='normal': change to 0.0
    attack = 1.0
    if line_split[41]=='normal.':
        attack = 0.0
        
    return LabeledPoint(attack, array([float(x) for x in clean_line_split]))

```

```python
#creating a new train/test data 
training_data = data.map(create_labeled_point)
testing_data = test_data.map(create_labeled_point)

#training_data.take(1)
testing_data.take(10)
```

## Feature selection

After we have prepared the data, we will train the model with all the features and select the important resluts based on its accuracy 


### Training decision tree classifier

```python
from pyspark.mllib.tree import DecisionTree, DecisionTreeModel
```

```python
#decision tree classifier 
tree_model = DecisionTree.trainClassifier(training_data, numClasses=2, 
                                          categoricalFeaturesInfo={1: len(network_protocols), 2: len(network_services), 3: len(network_flags)},
                                          impurity='gini', maxDepth=4, maxBins=100)
```

```python
tree_model
```

### Evaluating the model

```python
#get the predictions from the classification model
predictions = tree_model.predict(testing_data.map(lambda p: p.features))
#create a set of predictions and labels 
labels_and_preds = testing_data.map(lambda p: p.label).zip(predictions)
predictions.take(10)
labels_and_preds.take(40)
```

```python
#calculate the accuracy of the model by comparing the predicted values and the values in the testind_data
test_accuracy = labels_and_preds.filter(lambda v : v[0]==v[1] ).count()/ float(testing_data.count())
print ("test accuracy", round(test_accuracy,4))
```

## Building prediction model based on selected features

```python
def create_labeled_point_minimal(line_split):
    # leave_out = [41]
    clean_line_split = line_split[3:4]  + line_split[22:23] + line_split[33:38]
     
    # convert flag_variable to numeric categorical variable
    try:
        clean_line_split[0] = network_flags.index(clean_line_split[0])
    except:
        clean_line_split[0] = len(network_flags)
    
    # convert label to binary label
    attack = 1.0
    if line_split[41]=='normal.':
        attack = 0.0
        
    return LabeledPoint(attack, array([float(x) for x in clean_line_split]))

training_data_minimal = data.map(create_labeled_point_minimal)
test_data_minimal = test_data.map(create_labeled_point_minimal)
```

```python
training_data_minimal.take(1)
```

```python
# Build the model
tree_model_minimal = DecisionTree.trainClassifier(
    training_data_minimal, numClasses=2, 
    categoricalFeaturesInfo={ 0: len(network_flags)},
    impurity='gini', maxDepth=3, maxBins=32)
```

```python
predictions_minimal = tree_model_minimal.predict(test_data_minimal.map(lambda p: p.features))
labels_and_preds_minimal = test_data_minimal.map(lambda p: p.label).zip(predictions_minimal)

test_accuracy = labels_and_preds_minimal.filter(lambda v : v[0]==v[1] ).count() / float(test_data_minimal.count())
print(round(test_accuracy,4))
```
